###### Lexical Diversity

```python
def lexical_diversity(text):

    tokens_kylie = text.lower().split()

    return len(set(tokens_kylie))/len(tokens_kylie)

 

print("lexical diversity :",lexical_diversity(tweet.text))


def lexical_diversity(text):

    tokens_hudabeauty = text.lower().split()

    return len(set(tokens_hudabeauty))/len(tokens_hudabeauty)

 

print("lexical diversity :",lexical_diversity(tweet.text))



def lexical_diversity(text):

    tokens_rana = text.lower().split()

    return len(set(tokens_rana))/len(tokens_rana)

 

print("lexical diversity :",lexical_diversity(tweet.text))


```

```python
import pickle
from nltk_helpers import get_sentiments

with open('kylie_tweets.pkl','rb') as f:
    tweets=pickle.load(f)
print("\n --------------------KYLIE JENNER----------------------- \n")
#print(tweets)
print(get_sentiments(tweets[0].text))
print(get_sentiments(tweets[0].text))
print(get_sentiments(tweets[1].text))
for tweet in tweets:
    print(get_sentiments(tweet.text))

import pickle
from nltk_helpers import get_sentiments

with open('RanaDaggubati_tweets.pkl','rb') as f:
    tweets=pickle.load(f)
print("\n --------------------RANA DAGGUBATI----------------------- \n")
#print(tweets)
print(get_sentiments(tweets[0].text))
print(get_sentiments(tweets[0].text))
print(get_sentiments(tweets[1].text))
for tweet in tweets:
    print(get_sentiments(tweet.text))
    
import pickle
from nltk_helpers import get_sentiments

with open('hudabeauty_tweets.pkl','rb') as f:
    tweets=pickle.load(f)
print("\n --------------------HUDA BEAUTY----------------------- \n")
#print(tweets)
print(get_sentiments(tweets[0].text))
print(get_sentiments(tweets[0].text))
print(get_sentiments(tweets[1].text))
for tweet in tweets:
    print(get_sentiments(tweet.text))
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

#build our figure
fig = plt.figure()
# build the axes
ax = fig.add_subplot(111, projection='3d')

sentiments = [get_sentiments(tweet.text) for tweet in tweets]

# build our xs, ys, and zs
xs = []
for sent in sentiments:
    xs.append(sent['neg'])

ys = [sent['neu'] for sent in sentiments]
zs = list(map(lambda x: x['pos'], sentiments))

ax.scatter(xs, ys, zs)

#add another set of data
import pickle
with open('kylie_tweets.pkl', 'rb') as f:
    kylie_tweets = pickle.load(f)

from nltk_helpers import split_sentiments

kylieX, kylieY, kylieZ = split_sentiments([get_sentiments(tweet.text) for tweet in kylie_tweets])
ax.scatter(kylieX, kylieY, kylieZ, color='r', marker='^')

ax.set_xlabel('Negative')
ax.set_ylabel('Neutral')
ax.set_ylabel('Positive')
plt.title('Kylie Jenner')
plt.show()



from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

#build our figure
fig = plt.figure()
# build the axes
ax = fig.add_subplot(111, projection='3d')

sentiments = [get_sentiments(tweet.text) for tweet in tweets]

# build our xs, ys, and zs
xs = []
for sent in sentiments:
    xs.append(sent['neg'])

ys = [sent['neu'] for sent in sentiments]
zs = list(map(lambda x: x['pos'], sentiments))

ax.scatter(xs, ys, zs)

#add another set of data
import pickle
with open('hudabeauty_tweets.pkl', 'rb') as f:
    hudabeauty_tweets = pickle.load(f)

from nltk_helpers import split_sentiments

hudabeautyX, hudabeautyY, hudabeautyZ = split_sentiments([get_sentiments(tweet.text) for tweet in hudabeauty_tweets])
ax.scatter(hudabeautyX, hudabeautyY, hudabeautyZ, color='r', marker='^')

ax.set_xlabel('Negative')
ax.set_ylabel('Neutral')
ax.set_ylabel('Positive')
plt.title('HudaBeauty')
plt.show()



from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

#build our figure
fig = plt.figure()
# build the axes
ax = fig.add_subplot(111, projection='3d')

sentiments = [get_sentiments(tweet.text) for tweet in tweets]

# build our xs, ys, and zs
xs = []
for sent in sentiments:
    xs.append(sent['neg'])

ys = [sent['neu'] for sent in sentiments]
zs = list(map(lambda x: x['pos'], sentiments))

ax.scatter(xs, ys, zs)

#add another set of data
import pickle
with open('RanaDaggubati_tweets.pkl', 'rb') as f:
    RanaDaggubati_tweets = pickle.load(f)

from nltk_helpers import split_sentiments

RanaDaggubatiX, RanaDaggubatiY, RanaDaggubatiZ = split_sentiments([get_sentiments(tweet.text) for tweet in RanaDaggubati_tweets])
ax.scatter(RanaDaggubatiX, RanaDaggubatiY, RanaDaggubatiZ, color='r', marker='^')

ax.set_xlabel('Negative')
ax.set_ylabel('Neutral')
ax.set_ylabel('Positive')
plt.title('Rana Daggubati')
plt.show()



plt.scatter(xs,zs,label='kylie')
plt.scatter(hudabeautyX,hudabeautyZ,color='r',marker='$\clubsuit$',label='hudabeauty')
plt.scatter(RanaDaggubatiX,RanaDaggubatiZ,color='g',marker='d',label='RanaDaggubati')
plt.xlabel('negative')
plt.ylabel('positive')
plt.title('positive vs negative Sentiments')
plt.legend(loc='upper right',shadow=True)
plt.show()
```
